#### Re-Packaging and Reduction
Here we reduce the training data set so that we can have short computation times and also re-package the data so that it can be used by `sklearn`.

````python
from sklearn.model_selection import train_test_split

#Re-package data
Xshape = trainX.shape
trainX = np.reshape(trainX, (Xshape[0], Xshape[1]*Xshape[2]))
Xshape = testX.shape
testX = np.reshape(testX, (Xshape[0], Xshape[1]*Xshape[2]))

#Reduce training dataset
trainX_reduced, trainy_reduced = train_test_split(trainX, trainy, train_size = .05)[::2]
````

## 2. Pre-analysis: 

Using dimensionality reduction (for example principle component analysis) show graphically by example how digits cluster.

Questions you might ask: Which digits are better separated than others?

We will use principle component anlysis to show how digits cluster graphically. We will reduce to three dimensions because this is the most dimensions we can reasonably visualize. 

````python
from sklearn.decomposition import PCA
from scipy.stats import gaussian_kde
from mpl_toolkits.mplot3d import Axes3D

#Do PCA to reduce dimensionality to one
pca = PCA(3)
pca.fit(trainX)
trainX_3D = pca.transform(trainX)

#Make colors
colors = []
rng = np.linspace(0, .9, 10)
cmap = mpl.cm.get_cmap('hsv')
for i in rng:
    colors.append(cmap(i))

#We now visualize results
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
for i in range(10):
    ind = np.where(trainy == i)
    sing_targ = trainX_3D[ind]
    ax.plot(sing_targ[:, 0], sing_targ[:, 1], zs = sing_targ[:, 2], marker = '.',
            c = colors[i], ms = 5, label = str(i),
           ls = ' ')
plt.legend(loc = 'upper left')
ax.set_title('Data Reduced to 3 Dimensions')
ax.set_xlabel('Principle Component 1')
ax.set_ylabel('Principle Component 2')
ax.set_zlabel('Principle Component 3')
fig.set_size_inches(12, 6);
````

From this it does not look like the data separates well. The data with a target of 0 seems to separate the best. We will try a visualization that reduces the dataset to one dimension now.

````python
from scipy.stats import gaussian_kde

#Do PCA to reduce the dimensionality
pca = PCA(1)
trainX_pca = pca.fit_transform(trainX)

#We now visualize results
fig = plt.figure()
ax = fig.add_subplot(111)
for i in range(10):
    ind = np.where(trainy == i)
    sing_targ = trainX_pca[ind]
    lam = gaussian_kde(sing_targ.flatten())(sing_targ.flatten())
    idx = lam.argsort()
    sing_targ, lam = sing_targ[idx], lam[idx]
    ax.scatter(sing_targ.flatten(), np.zeros_like(sing_targ) + i,
               c = lam)
ax.tick_params(labelbottom = False, bottom = False) 
ax.set_yticks(np.arange(10))
ax.set_title('Data Reduced to One Dimension');
````

The above plot reduces show the data with its dimensionality reduced to one, separated by target and with color denoting the linear density with lighter colors being more dense. It is clear that 1 and 0 separate best.

We start by scaling our data with `sklearn.preprocessing.StandardScaler`.

````python
from sklearn.preprocessing import StandardScaler

#scale data
scaler = StandardScaler()
trainX_scaled = scaler.fit_transform(trainX_reduced*1.,
                                trainy_reduced)
testX_scaled = scaler.fit_transform(testX*1., testy)
````

We will try three different kernels for our SVMs, gaussians, polynomials, and sigmoids. We begin by making a function that will use SVM on our dataset given a kernel and parameter grid for grid search cross validation.

````python
from sklearn.svm import SVC
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import GridSearchCV
from time import time

def svc_fc(kernel, param_grid, cv) :
    """Performs a grid search cross validation using param_grid and then
    uses a SVM to fit trainX_reduced. Returns the grid search object that 
    sklearn.model_selection.GridSearchCV returns.
    Parameters
    ----------
    kernel : (str) name of the kernel to be used by sklearn.svm.SVC
    param_grid : (dictionary) to be passed to param_grid in
                 sklearn.model_selection.GridSearchCV
    cv : (int) how many folds to have in the grid search cross validation
    """
    #Choose available amount of RAM for computations in MB
    RAM = 9000

    #Make our model which does PCA then uses a SVM
    pca = PCA(n_components = 100, whiten = True)
    svc = SVC(kernel = kernel, cache_size = RAM, gamma = 'scale')
    model = make_pipeline(pca, svc)

    #Do a grid search to find the best parameters
    gd_sr = GridSearchCV(estimator = model,
                         param_grid = param_grid,
                         scoring = 'accuracy',
                         n_jobs = -1,
                         cv = cv,
                         iid = False)

    #Time how long the grid search takes
    t0 = time()
    gd_sr.fit(trainX_scaled, trainy_reduced)
    t = time() - t0
    print('The grid search took {0} s.'.format(t))
    print('The best found parameters are {0}.'.format(gd_sr.best_params_))
    return gd_sr
````

We now try using a SVM with a gaussian kernel. We will only use two cross folds to maximize the amount of data we can use.

````python
gd_sr = svc_fc("rbf",
              {'svc__C': [1000, 5000, 10000, 50000],
             'svc__gamma': [.005, .01, .05, .1]},
              2)
````

The best parameters are in the middle of our grid which is great! We will now see how well our model fits the test data. We will also make a confusion matrix and classification report. We will make a function that will do all of this for future use.

````python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
import seaborn as sns

#Make function for characterizing predictions
def pred_char(gd_sr) :
    """Makes prediction and characterizes results. Returns nothing.
    Parameters
    ----------
    gd_sr : (dictionary-like) object retured by sklearn.model_selection.GridSearchCV
    """
    model = gd_sr.best_estimator_
    y_pred = model.predict(testX_scaled)
    acc = np.sum(testy == y_pred)/y_pred.size * 100
    print('The fit had an accuracy of {0}% in predicting the test data.'.format(acc))
    target_names = []
    for i in range(10):
        target_names.append(str(i))
    print("Classification report:\n%s\n"
      % (classification_report(testy, y_pred)))
    print(classification_report(testy, y_pred))
    mat = confusion_matrix(testy, y_pred)
    fig, ax = plt.subplots(figsize = (9, 9))
    sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False,
            xticklabels=target_names,
            yticklabels=target_names)
    plt.xlabel('true label')
    plt.ylabel('predicted label')
    plt.title('Confusion Matrix');
    return None  

#Characterize predictions
pred_char(gd_sr)
````

Now we try a polynomial kernel.

````python
gd_sr = svc_fc("poly",
              {"svc__C" : [.1, .5, 1, 5],
              "svc__degree": [2, 3, 4, 5],
              "svc__coef0" : [-1, 0, 1, 2]},
              2)
````

The best parameters are in the middle of the grid. We now characterize predictions.

````python
pred_char(gd_sr)
````

Now we try a sigmoid kernel.

````python
gd_sr = svc_fc("sigmoid",
              {"svc__C" : [.1, .5, 1, 5],
              "svc__coef0" : [.01, .05, .1, .5]},
               2)
````

The best parameters are in the middle of the grid. We now characterize predictions.

````python
pred_char(gd_sr)
````

The model using the polynomial kernel predicted the highest percetnage of the test data correctly. ## 4. Use any other machine learning method (discussed in class or otherwise) and apply it to this problem. How does it compare to SVM?

We will try using a forest. We will use `sklearn.ensemble.RandomForestClassifier`.

````python
#Import the random forest classifier
from sklearn.ensemble import RandomForestClassifier

#Make random forest classifier
rfc = RandomForestClassifier(n_estimators = 100)

#We time how long it takes to fit data
t0 = time()
rfc.fit(trainX, trainy)
t = time() - t0
print("The fit took {0} s.".format(t))

#Make prediction on test data
y_pred = rfc.predict(testX)

#Calculate the accuracy of the fit
acc = np.sum(testy == y_pred)/y_pred.size * 100
print('The fit had an accuracy of {0}% in predicting the test data.'.format(acc))
target_names = []
for i in range(10):
    target_names.append(str(i))
print("Classification report:\n%s\n"
% (classification_report(testy, y_pred)))
print(classification_report(testy, y_pred))
mat = confusion_matrix(testy, y_pred)
fig, ax = plt.subplots(figsize = (9, 9))
sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False,
        xticklabels=target_names,
        yticklabels=target_names)
plt.xlabel('true label')
plt.ylabel('predicted label')
plt.title('Confusion Matrix');
````

The random forest classifier was able to use the entire training dataset in a reasonable amount of time and then make reasonably accurate predictions. SVM will have more accurate predictions but will take a very long time to use the entire training dataset. The random forest classifier predicted the test data with an accuracy very close to that of the polynomical kernel SVM. The random forest classifier seems like the most practical choice here.
